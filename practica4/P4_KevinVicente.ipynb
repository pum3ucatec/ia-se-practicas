## Cargar el texto del libro

```python
url = "https://www.gutenberg.org/files/11/11-0.txt"  # Libro: Alicia en el país de las maravillas
response = requests.get(url)
texto = response.text
```

#Tokenización del texto

```python
import nltk

# Descargar el recurso 'punkt'
nltk.download('punkt')

# Tokenización del texto
tokens = nltk.word_tokenize(texto)
print(tokens[:10])  # Mostrar los primeros tokens para verificar
```

## Generar n-gramas

```python
n = 4
ngramas = list(ngrams(tokens, n))
```

# Tareas

# Tarea 1: Cambiar la longitud de los n-gramas a 4 o 5

```python
longitud_ngramas = 4  # Cambia a 4 o 5 según la tarea
ngramas = list(ngrams(tokens, longitud_ngramas))

# ===== Mostrar los primeros 10 n-gramas =====
print(f"Primeros 10 n-gramas de longitud {longitud_ngramas}:")
print(ngramas[:10])

longitud_ngramas = 5  # Cambia a 4 o 5 según la tarea
ngramas = list(ngrams(tokens, longitud_ngramas))

# ===== Mostrar los primeros 10 n-gramas =====
print(f"Primeros 10 n-gramas de longitud {longitud_ngramas}:")
print(ngramas[:10])
```

# Tarea 2: Utiliza la función FreqDist de NLTK para calcular la frecuencia de cada n-grama.

```python
freq_ngramas = FreqDist(ngramas)

# Mostrar los 10 n-gramas más frecuentes
print("\nLos 10 n-gramas más frecuentes:")
for ngrama, frecuencia in freq_ngramas.most_common(10):
    print(" ".join(ngrama), "-", frecuencia)
```

# Tarea 3: Utiliza la función plot de matplotlib para visualizar la distribución de frecuencias de los n-gramas.

```python
ngramas_más_frecuentes = freq_ngramas.most_common(10)
etiquetas = [" ".join(ng[0]) for ng in ngramas_más_frecuentes]
frecuencias = [ng[1] for ng in ngramas_más_frecuentes]

# Crear la gráfica de barras
plt.figure(figsize=(12, 6))
plt.bar(etiquetas, frecuencias, color='skyblue')
plt.xticks(rotation=45, ha="right")
plt.xlabel(f"N-gramas (longitud {longitud_ngramas})")
plt.ylabel("Frecuencia")
plt.title(f"Distribución de Frecuencias de los {longitud_ngramas}-gramas")
plt.tight_layout()
plt.show()
```
